```python
# Task 1 — Split data and train models

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score

# Load data (as provided)
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification

X, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=6,
    n_redundant=2,
    random_state=42
)
df = pd.DataFrame(X, columns=[f"feature_{i+1}" for i in range(10)])
df["default"] = y

# Split into 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(
    df.drop("default", axis=1),
    df["default"],
    test_size=0.2,
    random_state=42
)

# Train a single Decision Tree
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
y_train_pred_tree = tree.predict(X_train)
y_test_pred_tree = tree.predict(X_test)

# Train a BaggingClassifier with 100 decision trees
bag = BaggingClassifier(
    estimator=DecisionTreeClassifier(), # Changed from base_estimator
    n_estimators=100,
    bootstrap=True,
    oob_score=False,
    random_state=42
)
bag.fit(X_train, y_train)
y_train_pred_bag = bag.predict(X_train)
y_test_pred_bag = bag.predict(X_test)

# Print training and test accuracies
print("Decision Tree - Train Accuracy:", accuracy_score(y_train, y_train_pred_tree))
print("Decision Tree - Test  Accuracy:", accuracy_score(y_test,  y_test_pred_tree))
print("Bagging       - Train Accuracy:", accuracy_score(y_train, y_train_pred_bag))
print("Bagging       - Test  Accuracy:", accuracy_score(y_test,  y_test_pred_bag))


# Task 2 — OOB scoring

# Enable Out-of-Bag (OOB) scoring
bag_oob = BaggingClassifier(
    estimator=DecisionTreeClassifier(), # Changed from base_estimator
    n_estimators=100,
    bootstrap=True,
    oob_score=True,  # enable OOB
    random_state=42
)
bag_oob.fit(X_train, y_train)

# Compute OOB score
oob_score = bag_oob.oob_score_

print("Bagging OOB Score:", oob_score)

# The OOB score is the accuracy of the ensemble on the "out-of-bag" samples
# that were not used in the bootstrap for each estimator. It serves as a built-in
# cross-validation metric, useful for estimating generalisation without a separate validation set.


# Task 3 — Interpretation

"""
The BaggingClassifier generalizes better than the single decision tree.
The decision tree achieved 100% accuracy on the training set (overfitting)
but only ~82% on the test set. The bagged ensemble also fits the training data
(100% accuracy) but achieves a higher test accuracy (~87%) and a high OOB score (~90%).
This indicates that the ensemble is more robust.

Bagging reduces variance by averaging multiple trees trained on different bootstrap samples.
Each tree captures different patterns (and different errors) in the data, and their aggregate
prediction is less sensitive to noise and overfitting than a single tree. As a result,
the bagged model has lower variance and better generalisation on unseen data.
"""
```
