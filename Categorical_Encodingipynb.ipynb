```python
import pandas as pd

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, RobustScaler

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)

# ── Task 1A: Label Encoding — Gender ─────────────────────────────────────────
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
# ── Task 1B: One-Hot Encoding — City ─────────────────────────────────────────
df = pd.get_dummies(df, columns=['City'], drop_first=True)
# ── Task 1C: Ordinal Label Encoding — Size ───────────────────────────────────
size_order = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_order)


# ── Task 2: Feature Scaling — Age and Fare ───────────────────────────────────
scaler = RobustScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
print("── Encoded & Scaled DataFrame ──────────────────")
print(df.to_string(index=True))
print()
print("── Column dtypes ────────────────────────────────")
print(df.dtypes)
```
