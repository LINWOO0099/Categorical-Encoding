```python
import csv
import json
import io


def csv_to_json(csv_string: str) -> str:

    # Wrap CSV string using StringIO
    csv_file = io.StringIO(csv_string)


    # Create CSV reader
    reader = csv.reader(csv_file)


    # Skip header row
    next(reader)


    data = []


    # Process each row
    for row in reader:

        record = {

            "order_id": row[0],

            "customer": row[1],

            "city": row[2],

            "amount": float(row[3]),

            "delivered": True if row[4] == "True" else False

        }


        data.append(record)


    # Convert list to JSON string
    return json.dumps(data)



# -----------------------------
# Sample Test
# -----------------------------

if __name__ == "__main__":


    sample = """order_id,customer,city,amount,delivered
O001,Alice Corp,Mumbai,1500.50,True
O002,Beta Ltd,Delhi,320.00,False
O003,Gamma Inc,Bangalore,875.75,True
O004,Delta Co,Chennai,210.00,False
O005,Epsilon LLC,Hyderabad,999.99,True"""


    print(csv_to_json(sample))
```
